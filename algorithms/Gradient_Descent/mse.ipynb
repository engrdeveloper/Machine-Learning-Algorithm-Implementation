```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Create tiny data
X = np.array([1, 2, 3, 4, 5])
y = np.array([-3, 7, 9, 11, 13])
w = np.zeros(2)
lr = 0.01 

# Add bias column of 1s
X_bias = np.c_[np.ones(X.shape[0]), X]

mse_losses = []
mae_losses = []

for epecho in range(1000):
    y_pred = X_bias.dot(w)
    error = y - y_pred
    
    mse = np.mean(error ** 2)
    mae = np.mean(np.abs(error))
    mse_losses.append(mse)
    mae_losses.append(mae)

    grad = -2 * X_bias.T.dot(error) / len(X)

    w -= lr * grad

print(f"Final weights: {w}")


# Plot the fitted line
plt.figure(figsize=(6,4))
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X, X_bias.dot(w), color='red', label='Fitted line (MSE)')
plt.legend()
plt.show()

# Plot MSE vs MAE losses
plt.figure(figsize=(6,4))
plt.plot(mse_losses, label='MSE Loss')
plt.plot(mae_losses, label='MAE Loss', linestyle='--')
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("MSE vs MAE during training")
plt.legend()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Dummy Data
X = np.array([1, 2, 3, 4, 5])
y = np.array([-3, 7, 9, 11, 13])

# Add bias column
X_bias = np.c_[np.ones(X.shape[0]), X]

# Initialize weights
w = np.zeros(2)
lr = 0.01
epochs = 50  # fewer epochs, we update more frequently

# Store losses
mse_losses = []

for epoch in range(epochs):
    # Shuffle data each epoch for SGD
    indices = np.arange(len(X))
    np.random.shuffle(indices)
    
    for i in indices:
        xi = X_bias[i].reshape(1, -1)   # 1x2
        yi = y[i]
        
        # Prediction
        y_pred = xi.dot(w)
        error = yi - y_pred
        
        # Gradient (for this single sample)
        grad = -2 * xi.T.dot(error)
        
        # Update weights
        w -= lr * grad
    
    # Compute loss for monitoring (full batch)
    y_pred_all = X_bias.dot(w)
    loss = np.mean((y - y_pred_all) ** 2)
    mse_losses.append(loss)

print(f"Final weights: {w}")

# Plot fitted line
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X, X_bias.dot(w), color='red', label='Fitted line (SGD)')
plt.legend()
plt.show()

# Plot loss curve
plt.figure(figsize=(6,4))
plt.plot(mse_losses)
plt.title("MSE Loss Curve - SGD")
plt.xlabel("Epochs")
plt.ylabel("MSE")
plt.show()
```
